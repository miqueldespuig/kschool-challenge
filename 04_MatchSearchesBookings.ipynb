# Match Searches with Bookings

* For every search in the searches file, find out whether
the search ended up in a booking or not (using the info
in the bookings file). For instance, search and booking
origin and destination should match.
* For the bookings file, origin and destination are the
columns dep_port and arr_port, respectively.
* Generate a CSV file with the search data, and an
additional field, containing 1 if the search ended up in
a booking, and 0 otherwise.

# Modules and libraries

```python
import pandas as pd
```

```python
path = '/home/miki/Documents/data/challenge/'
f_searches = 'searches.csv.bz2'
f_bookings = 'bookings.csv.bz2'
```

# Remove duplicates

Start with the smaller file. Toy, toy, toy problem

```python
df_bookings = pd.read_csv(
    path+f_bookings,
    nrows = 1e4,
    sep = '^'    
)
df_bookings.head(5)
```

```python
df_bookings.drop_duplicates(inplace=True)
df_bookings
```

Inplace, to save some memory starting now with the iterator.

```python
nrows = 1e5
chunksize = 1e6
bookings_iter = pd.read_csv(
    path+f_bookings,
    delimiter = '^',
    dtype = str,
    #nrows = nrows,
    chunksize= chunksize
)
df_bookings = pd.DataFrame()
for i,chunk in enumerate(bookings_iter):
    df_bookings = df_bookings.append(chunk)
    df_bookings.drop_duplicates(inplace=True)
    print(i, len(df_bookings))
```

```python
df_bookings.to_csv(path+'bookings_no_dup.csv.bz2',sep='^',compression='bz2')
```

```python
chunksize = 500000
searches_iter = pd.read_csv(
    path+f_searches,
    sep = '^',
    dtype = str,
    memory_map = True,
    #nrows = nrows,
    chunksize= chunksize
)
df_searches = pd.DataFrame()
for chunk in searches_iter:
    df_searches = df_searches.append(chunk)
    df_searches.drop_duplicates(inplace=True)
```

```python
df_searches.to_csv(path+'searches_no_dup.csv.bz2',sep='^',compression='bz2')
```
